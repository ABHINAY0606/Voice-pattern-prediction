```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
df = pd.read_csv("/content/features_with_clusters.csv")
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x="cluster", palette="Set2")
plt.title("Distribution of Cognitive Risk Clusters")
plt.xlabel("Cluster (Risk Group)")
plt.ylabel("Number of Samples")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[["pauses", "hesitations", "speech_rate", "pitch_var"]].corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df, vars=["pauses", "hesitations", "speech_rate", "pitch_var"], hue="cluster", palette="Set1")
plt.suptitle("Feature Relationships Colored by Cluster", y=1.02)
plt.show()
```

```python
features = ["pauses", "hesitations", "speech_rate", "pitch_var"]

for feature in features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x="cluster", y=feature, palette="pastel")
    plt.title(f"{feature.title()} by Cluster")
    plt.show()
```

```python
subset = df.head(3)
features = ["pauses", "hesitations", "speech_rate", "pitch_var"]
angles = np.linspace(0, 2 * np.pi, len(features), endpoint=False).tolist()
angles += angles[:1]

plt.figure(figsize=(6, 6))
for i, row in subset.iterrows():
    values = row[features].tolist()
    values += values[:1]
    plt.polar(angles, values, label=row["filename"])

plt.xticks(angles[:-1], features)
plt.title("Radar Chart: Feature Profiles Across Samples")
plt.legend(loc="upper right", bbox_to_anchor=(1.3, 1))
plt.show()
```

```python
print("\nMost indicative features based on correlation and visualization:")
print("- High number of pauses and hesitations likely cluster with higher risk.")
print("- Lower speech rate may indicate cognitive issues.")
print("- Pitch variation provides some insight, but less strongly.")

print("\nSuggested next steps:")
print("1. Analyze cluster characteristics more deeply (e.g., average feature values).")
print("2. Map clusters to known cognitive risk levels if available.")
print("3. Expand dataset and validate using clinical metadata if possible.")
```
